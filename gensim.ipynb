```python
words=[]
from nltk.tokenize import WordPunctTokenizer
for sentence in sentences:
    word = WordPunctTokenizer().tokenize(sentence)
    words.append(word)
print(words)
```

```python
words1=words
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
num_sentences=len(words)
for i in range(num_sentences):
    num_words_perst=len(words[i])
    for j in range(num_words_perst):
        word=words[i][j]
        words[i][j]=lemmatizer.lemmatize(word)
print(words)
```

```python
#另外一种清理数据的方法
#https://www.analyticsvidhya.com/blog/2016/08/beginners-guide-to-topic-modeling-in-python/
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

sentences_clean = [clean(doc).split() for doc in sentences] 
print(sentences_clean)
```

```python
#word2vec model
w2v_model = gensim.models.Word2Vec(sentences_clean, min_count=1)
```

```python
from gensim import corpora
#create dictionary
dictionary=corpora.Dictionary(sentences_clean)
```

```python
print(dictionary)
```

```python
# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above. 
term_matrix = [dictionary.doc2bow(sentence) for sentence in sentences_clean]
```

```python
# Creating the object for LDA model using gensim library
lda = gensim.models.ldamodel.LdaModel
```

```python
# Running and Training LDA model on the document term matrix
ldamodel = lda(term_matrix, num_topics=3, id2word = dictionary, passes=50)
```

```python
# Results 
print(ldamodel.print_topics())
```

```python
#一共有3600个词和100个维度
weight_matrix=np.zeros((3600,100))
for i in range(3600):
    token=dictionary[i]
    weight_matrix[i,]=w2v_model[token]
print(weight_matrix)
```

```python
#我们现在要一个反转的dictionary
def invert_dict(d):  
    return dict((v,k) for k,v in d.items()) 

dictionary_inverse=invert_dict(dictionary)
```

```python
import pandas as pd
weight_matrix_frame=pd.DataFrame(weight_matrix,index=dictionary_inverse)
print(weight_matrix_frame)
```
